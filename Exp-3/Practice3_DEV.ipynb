Q1. Using the data below, create a bubble plot where the x-axis represents the GDP (in trillion  USD), the y-axis represents the population (in millions), and the size of the bubble represents  the carbon emissions (in million tons): Country A: GDP - 3, Population - 150, Emissions - 500; Country B: GDP - 5, Population - 200, Emissions - 800;  
Country C: GDP - 2, Population - 100, Emissions - 300; Country D: GDP - 4,  Population - 250, Emissions - 600. What insights can you gain from the bubble plot?

```python
import matplotlib.pyplot as plt

countries = ['Country A', 'Country B', 'Country C', 'Country D']
gdp = [3, 5, 2, 4]
population = [150, 200, 100, 250]
emissions = [500, 800, 300, 600]
plt.figure(figsize=(10, 6))
bubble = plt.scatter(gdp, population, s=[e / 2 for e in emissions], alpha=0.6, c=emissions, cmap='viridis', edgecolors='w', linewidth=2)

plt.colorbar(bubble, label='Carbon Emissions (Million Tons)')

plt.xlabel('GDP (Trillion USD)')
plt.ylabel('Population (Millions)')
plt.title('Bubble Plot of GDP vs. Population with Emissions as Bubble Size')

for i, country in enumerate(countries):
    plt.text(gdp[i], population[i], country, fontsize=12, ha='right')

plt.grid(True)
plt.show()
```

Q2.  2. Given the following data for five products and their respective sales in units: Product A: 120,  Product B: 180, Product C: 140, Product D: 200, Product E: 160, create a bar chart  to represent the data. Which product had the highest and lowest sales?


```python
import matplotlib.pyplot as plt

products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [120, 180, 140, 200, 160]

plt.figure(figsize=(10, 6))
plt.bar(products, sales, color='skyblue')
plt.xlabel('Products')
plt.ylabel('Sales (Units)')
plt.title('Sales of Different Products')
plt.ylim(0, max(sales) + 20)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

highest_sales = max(sales)
lowest_sales = min(sales)
highest_sales_product = products[sales.index(highest_sales)]
lowest_sales_product = products[sales.index(lowest_sales)]

print(f"Product with the highest sales: {highest_sales_product} ({highest_sales} units)")
print(f"Product with the lowest sales: {lowest_sales_product} ({lowest_sales} units)")
```

Q3.  A company’s revenue is divided into four categories: Product Sales: 40%, Services: 30%,  Investments: 20%, Other: 10%. Create a pie chart to represent the revenue distribution.  How well does the pie chart represent the proportion of revenue from each category?


```python
import matplotlib.pyplot as plt

categories = ['Product Sales', 'Services', 'Investments', 'Other']
percentages = [40, 30, 20, 10]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=categories, colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.4))

plt.title('Revenue Distribution by Category')

plt.show()
```

Q4. Generate a histogram using random normal data points (mean = 0, standard deviation = 1) and  describe the shape of the distribution.


```python
import numpy as np
import matplotlib.pyplot as plt
mean = 0
std_dev = 1
num_points = 1000
data = np.random.normal(mean, std_dev, num_points)

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, edgecolor='k', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Normal Data Points (Mean = 0, Std Dev = 1)')
plt.grid(True)
plt.show()
```
