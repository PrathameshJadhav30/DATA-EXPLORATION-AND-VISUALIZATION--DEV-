9. Perform Time Series Analysis on sample data points and plot the Time Series

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate Sample Time Series Data
# ---------------------------------
np.random.seed(42)
date_range = pd.date_range(start='2024-01-01', periods=100, freq='D')  # 100 days starting from Jan 1, 2024
data = np.cumsum(np.random.randn(100) * 2) + 50  # Simulated cumulative data with random noise
```

```python
# Create a DataFrame
time_series_df = pd.DataFrame({'Date': date_range, 'Value': data})
time_series_df.set_index('Date', inplace=True)
```

```python
# 1. Plot the Time Series
# ---------------------------------
plt.figure(figsize=(12, 6))
sns.lineplot(x=time_series_df.index, y=time_series_df['Value'], label='Time Series Data')
plt.title('Time Series Plot', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```

```python
# 2. Decompose Time Series (Optional)
# ---------------------------------
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(time_series_df['Value'], model='additive', period=30)
decomposition.plot()
plt.show()
```

```python
# 3. Rolling Statistics for Trend Detection
# ---------------------------------
time_series_df['Rolling Mean'] = time_series_df['Value'].rolling(window=7).mean()  # 7-day rolling average
```

```python
plt.figure(figsize=(12, 6))
plt.plot(time_series_df['Value'], label='Original Data', alpha=0.5)
plt.plot(time_series_df['Rolling Mean'], label='7-Day Rolling Mean', color='red', linewidth=2)
plt.title('Time Series with Rolling Mean', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```
