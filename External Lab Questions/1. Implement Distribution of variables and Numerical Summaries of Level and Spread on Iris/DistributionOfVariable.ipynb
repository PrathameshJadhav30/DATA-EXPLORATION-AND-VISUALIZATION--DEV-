1. Implement Distribution of variables and Numerical Summaries of Level and Spread on Iris 
dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

```python
# Distribution of variables
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))
```

```python
# Plot histograms
for i, column in enumerate(df.columns[:-1]):
    plt.subplot(2, 2, i+1)
    sns.histplot(df[column], kde=True, color='blue', bins=20)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```

```python
# Pairplot
sns.pairplot(df, hue='species', diag_kind="kde", palette="husl")
plt.show()
```

```python
# Numerical Summaries of Level and Spread
summary = df.describe()  # Summary statistics
print("Summary Statistics:")
print(summary)
```

```python
# Compute IQR
iqr = df[df.columns[:-1]].apply(lambda x: x.quantile(0.75) - x.quantile(0.25))
print("\nInterquartile Range (IQR):")
print(iqr)
```

```python
# Additional statistics (mode)
mode = df[df.columns[:-1]].mode().iloc[0]
print("\nMode:")
print(mode)
```
