4. Implement Histogram, Bar Charts & Pie Charts on sample data points.

```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Generate sample data
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=500)  # Normally distributed data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 20, 15]
```

```python
# 1. Histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Sample Data', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# 2. Bar Chart
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=['red', 'green', 'blue', 'purple'], alpha=0.7)
plt.title('Bar Chart of Categories', fontsize=14)
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
# 3. Pie Chart
plt.figure(figsize=(8, 6))
explode = [0.1, 0, 0, 0]  # Highlight Category A
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90,
        colors=['red', 'green', 'blue', 'purple'], explode=explode, shadow=True)
plt.title('Pie Chart of Categories', fontsize=14)
plt.show()
```
