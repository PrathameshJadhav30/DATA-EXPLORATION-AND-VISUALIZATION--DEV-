11. Visualize hierarchical data structures using Dendrograms with sample data points

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Step 3: Generate Sample Data Points
np.random.seed(42)
data = np.random.rand(10, 2)  # 10 points in 2D space
```

```python
# Step 4: Perform Hierarchical Clustering
# 'linkage' computes the hierarchical clustering
linked = linkage(data, method='ward')  # 'ward' minimizes the variance of clusters
```

```python
# Step 5: Plot the Dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linked,
           orientation='top',
           labels=range(1, 11),  # Label each point from 1 to 10
           distance_sort='descending',
           show_leaf_counts=True)
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()
```
