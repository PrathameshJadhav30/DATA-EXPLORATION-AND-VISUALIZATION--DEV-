3. Implement Line Plots & Area Plots on sample data points.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Create sample data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y1 = np.sin(x)  # Sine function
y2 = np.cos(x)  # Cosine function
```

```python
# Create a DataFrame for easier plotting
data = pd.DataFrame({'x': x, 'sin(x)': y1, 'cos(x)': y2})
```

```python
# Line Plot
plt.figure(figsize=(10, 6))
plt.plot(data['x'], data['sin(x)'], label='sin(x)', color='blue', linestyle='-', linewidth=2)
plt.plot(data['x'], data['cos(x)'], label='cos(x)', color='orange', linestyle='--', linewidth=2)
plt.title('Line Plot of sin(x) and cos(x)', fontsize=14)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Area Plot
plt.figure(figsize=(10, 6))
plt.fill_between(data['x'], data['sin(x)'], color='blue', alpha=0.3, label='sin(x)')
plt.fill_between(data['x'], data['cos(x)'], color='orange', alpha=0.3, label='cos(x)')
plt.title('Area Plot of sin(x) and cos(x)', fontsize=14)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
