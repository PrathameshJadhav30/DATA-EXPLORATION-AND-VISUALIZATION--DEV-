10. Perform Time Series Analysis on sample data points and Decompose the Time Series into 
Trend, Seasonality and Residuals

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Step 3: Create Sample Time Series Data
np.random.seed(42)
date_range = pd.date_range(start='2020-01-01', periods=100, freq='D')
trend = np.linspace(50, 150, 100)
seasonality = 10 * np.sin(np.linspace(0, 2 * np.pi, 100))
noise = np.random.normal(0, 5, 100)
data = trend + seasonality + noise
```

```python
# Create a DataFrame
time_series_data = pd.DataFrame({'Date': date_range, 'Value': data})
time_series_data.set_index('Date', inplace=True)
```

```python
# Step 4: Decompose the Time Series
decomposition = seasonal_decompose(time_series_data['Value'], model='additive', period=30)
```

```python
# Step 5: Plot the Decomposition
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(411)
plt.plot(time_series_data['Value'], label='Original Time Series')
plt.legend(loc='best')
```

```python
plt.subplot(412)
plt.plot(decomposition.trend, label='Trend', color='orange')
plt.legend(loc='best')
```

```python
plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonality', color='green')
plt.legend(loc='best')
```

```python
plt.subplot(414)
plt.plot(decomposition.resid, label='Residuals', color='red')
plt.legend(loc='best')
```

```python
plt.tight_layout()
plt.show()
```
