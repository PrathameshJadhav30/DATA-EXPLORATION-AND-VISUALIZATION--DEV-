 Q1. Plot histograms for each feature to analyze the distribution of the data. What do the shapes of  the histograms suggest about the distribution?


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
```

```python
years = [2000, 2001, 2002, 2003, 2004, 2005]
population = [1000, 1050, 1100, 1200, 1300, 1400]

df = pd.DataFrame({
    'Year': years,
    'Population': population,
})

for column in df.columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[column], kde=True, bins=30)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

Q2. Create boxplots for each feature in the Iris dataset. How do the boxplots help identify any  outliers in the dataset?


```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=iris.drop(columns='species'), orient='h')
plt.title('Boxplots of Iris Dataset Features')
plt.xlabel('Value')
plt.ylabel('Features')
plt.show()
```
